```python
# Generate random number between 14 and 70 total of 100
```

```python
import random
numbers = []
counter = 1
while counter <= 100:
    num = random.randint(14,70)
    numbers.append(num)
    counter +=1
print(numbers) 
```

```python
import math
squareroot =list(map(lambda x: (math.sqrt(x)),numbers)) 
print(squareroot)
```

```python
import math
lognum =list(map(lambda x: (math.log(x)), numbers))
print(lognum)
```

```python
oddlist =list(filter(lambda x: (x%2 !=0)), numbers)
print(oddlist)
```
